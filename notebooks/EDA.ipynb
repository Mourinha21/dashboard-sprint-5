```python
#Renomeando a coluna "Revenue growth"
dados_eua = dados_eua.rename(columns={'Revenue growth': 'Revenue growth (%)'})
dados_eua.columns
```

```python
#Alterando o tipo da coluna 'Profit(billions US$) de object para float'

dados_india['Profit(billions US$)'][30] = -0.4
dados_india['Profit(billions US$)'] = pd.to_numeric(dados_india['Profit(billions US$)'])
dados_india.info()
```

```python
#Alterando os dados de um índice para conseguir converter a coluna no dataframe dos dados das Filipinas
dados_filipinas['Profits(USD millions)'][24] = '12'
```

```python
# Alterando o tipo das colunas do dataframe das Filipinas

dados_filipinas['Revenue(USD millions)'] = (dados_filipinas['Revenue(USD millions)'].str.replace(",", ".").astype(float))
dados_filipinas['Profits(USD millions)'] = (dados_filipinas['Profits(USD millions)'].str.replace(",", ".").astype(float))
dados_filipinas['Employees'] = (dados_filipinas['Employees'].str.replace(",", "").astype(float))

dados_filipinas.info()
```

```python
#Ajustando e depois convertendo as colunas que não eram do tipo certo
dados_eua['Revenue (USD millions)'] = pd.to_numeric(dados_eua['Revenue (USD millions)'].str.replace(",", ".").astype(float))
dados_eua['Revenue growth (%)'] = pd.to_numeric(dados_eua['Revenue growth (%)'].str.replace("%", "").astype(float))
dados_eua['Employees'] = dados_eua['Employees'].str.replace(",", ".")
dados_eua['Employees'] = dados_eua['Employees'].str.replace(".", "")
dados_eua['Employees'] = pd.to_numeric(dados_eua['Employees'].astype(float))

dados_eua.head()
```

```python
#Verificando se há duplicatas nos dataframes

print(dados_europa.duplicated().sum())
print(dados_india.duplicated().sum())
print(dados_filipinas.duplicated().sum())
print(dados_eua.duplicated().sum())
```

```python
#Agrupando dados da Europa por país
receita_pais_europeu = (
    dados_europa
        .groupby(['Headquarters'])[['Revenue(US$ billions)']]
        .mean().reset_index()
)

receita_pais_europeu.sort_values(by='Revenue(US$ billions)', axis=0, ascending=True, inplace=True)
receita_pais_europeu.head()
```

```python
#Criando gráficos de dispersão dos dados dos países Europeus
fig = px.histogram(dados_europa, x='Revenue(US$ billions)', 
             title='Contagem de valores de receita das empresa européias em bilhões de US$', color_discrete_sequence=['indianred'])
fig.show()
```

```python
#Ordenando os dados do dataframe da Índia para a criação do gráfico
dados_india.sort_values(by='Forbes 2000 rank', axis=0, ascending=False, inplace=True)
dados_india.head(10)
```

```python
#Criando gráfico de dispersão Rank Forbes x Patrimônio em bilhões US$
fig = px.scatter(dados_india, x='Forbes 2000 rank', 
                 y='Assets(billions US$)', 
                 trendline='expanding', 
                 title='Patrimônio de empresas de acordo com o ranking da Forbes')
fig.show()
```

```python
#Atribuindo os dados ordenados do dataframe das Filipinas para a criação do gráfico
dados_ordenados_filipinas = dados_filipinas.sort_values(by='Employees', axis=0, ascending=True)

dados_ordenados_filipinas.head()
```

```python
#Criando um gráfico para comparar o lucro das empresas por n° de funcionários
fig = px.scatter(dados_ordenados_filipinas, x='Employees',
                   y='Profits(USD millions)', 
                   title='Lucro de empresas (por milhões de US$) pelo o número de funcionários (por mil)')
fig.show()
```

```python
#Atribuindo os dados dos EUA em uma nova variável para a criação do gráfico
crescimento_industria_eua = (dados_eua.groupby(['Industry'])[['Revenue growth (%)']].sum().reset_index())
crescimento_industria_eua.sort_values(by='Revenue growth (%)', ascending=True, inplace=True)
```

```python
#Criando o gráfico de área para comparar o crescimento entre empresas
fig = px.area(crescimento_industria_eua, x='Industry', y='Revenue growth (%)', title='Crescimento de receita das empresas estadunidenses por indústria')
fig.show()
```
